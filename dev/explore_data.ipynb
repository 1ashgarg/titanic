```python
# Quick function to explore categorical variables
def cat_data_explore(var_name, dataset):
    print('Counts')
    print('--------------------------------')
    print(dataset[var_name].value_counts())
    print('')
    print('Survival group count')
    print('--------------------------------')  
    print(dataset[['Survived', var_name, 'PassengerId']].groupby(['Survived', var_name]).count())
    print('')
    print('Missing Data')
    print('--------------------------------')  
    print('missing count = ' + str(len(dataset) - dataset[var_name].count()))
    print(str((1 - dataset[var_name].count()/len(dataset))*100) + '%')
    print('')
    print('Graph')
    print('--------------------------------')
    sns.countplot(x=var_name, data = dataset)
```

```python
cat_data_explore('Pclass', train_df)
```

```python
cat_data_explore('Sex', train_df)
```

```python
cat_data_explore('Embarked', train_df)
```
